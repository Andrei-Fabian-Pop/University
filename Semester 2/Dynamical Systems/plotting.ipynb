```python
t,k=var('t,k')
x=function('x')(t)
eq=diff(x,t)==k*x
desolve(eq,[x,t],ics=[0,1])
```

```python
sol(t,k)=desolve(eq,[x,t],ics=[0,1])
sol(t,k)
```

```python
sols=[plot(sol(t,k/20),t,-2,2,xmin=0,xmax=50) for k in [-40..40]]
```
